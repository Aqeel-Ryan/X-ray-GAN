```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Paths
data_dir = 'D:/work/A2M/MediGAN/images_001/images'  # Change this to your actual images directory
csv_file = 'D:/work/A2M/MediGAN/Data_Entry_2017/Data_Entry_2017.csv'  # Change this to your actual CSV file

# Load the CSV file
labels_df = pd.read_csv(csv_file)
```

```python
# EDA: Show some sample images
def show_sample_images(data_dir, labels_df, sample_size=5):
    sample_images = labels_df.sample(n=sample_size)
    plt.figure(figsize=(15, 5))
    for i, img_name in enumerate(sample_images['Image Index']):
        img_path = os.path.join(data_dir, img_name)
        img = Image.open(img_path).convert('L')  # Convert to grayscale
        plt.subplot(1, sample_size, i+1)
        plt.imshow(img, cmap='gray')
        plt.title(sample_images.iloc[i]['Finding Labels'])
        plt.axis('off')
    plt.show()

show_sample_images(data_dir, labels_df)
```

```python
# EDA: Basic statistics about the images
def calculate_image_stats(data_dir, labels_df, sample_size=100):
    widths, heights = [], []
    sample_images = labels_df.sample(n=sample_size)
    for img_name in sample_images['Image Index']:
        img_path = os.path.join(data_dir, img_name)
        with Image.open(img_path) as img:
            width, height = img.size
            widths.append(width)
            heights.append(height)
    print(f"Average width: {sum(widths) / len(widths)}")
    print(f"Average height: {sum(heights) / len(heights)}")
    print(f"Width: min={min(widths)}, max={max(widths)}")
    print(f"Height: min={min(heights)}, max={max(heights)}")

calculate_image_stats(data_dir, labels_df)
```

```python
# Custom Dataset Class
class XRayDataset(Dataset):
    def __init__(self, data_dir, labels_df, transform=None):
        self.data_dir = data_dir
        self.labels_df = labels_df
        self.transform = transform

    def __len__(self):
        return len(self.labels_df)

    def __getitem__(self, idx):
        img_name = os.path.join(self.data_dir, self.labels_df.iloc[idx, 0])
        image = Image.open(img_name).convert('L')  # Convert to grayscale
        if self.transform:
            image = self.transform(image)
        return image

# Image transformations
transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
```

```python
# Prepare Dataset and DataLoader
dataset = XRayDataset(data_dir=data_dir, labels_df=labels_df, transform=transform)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True, num_workers=4)
```

```python
# GAN Models
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(100, 256),
            nn.ReLU(True),
            nn.Linear(256, 512),
            nn.ReLU(True),
            nn.Linear(512, 1024),
            nn.ReLU(True),
            nn.Linear(1024, 64*64),
            nn.Tanh()
        )

    def forward(self, x):
        x = self.main(x)
        x = x.view(-1, 1, 64, 64)
        return x

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(64*64, 1024),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(1024, 512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = x.view(-1, 64*64)
        x = self.main(x)
        return x
    
    
# Initialize models
generator = Generator()
discriminator = Discriminator()
```

```python
# Loss and Optimizer
criterion = nn.BCELoss()
optimizer_g = optim.Adam(generator.parameters(), lr=0.0002)
optimizer_d = optim.Adam(discriminator.parameters(), lr=0.0002)
```

```python
# Early Stopping Parameters
patience = 6
best_loss = float('inf')
trigger_times = 0

# Save path
save_path_g = 'generator.pth'
save_path_d = 'discriminator.pth'

# Training Loop
num_epochs = 70
for epoch in range(num_epochs):
    for i, data in enumerate(dataloader):
        # Train Discriminator
        real_data = data
        batch_size = real_data.size(0)
        labels_real = torch.ones(batch_size, 1)
        labels_fake = torch.zeros(batch_size, 1)

        optimizer_d.zero_grad()
        output_real = discriminator(real_data).view(-1, 1)
        loss_real = criterion(output_real, labels_real)
        
        noise = torch.randn(batch_size, 100)
        fake_data = generator(noise)
        output_fake = discriminator(fake_data.detach()).view(-1, 1)
        loss_fake = criterion(output_fake, labels_fake)
        
        loss_d = loss_real + loss_fake
        loss_d.backward()
        optimizer_d.step()

        # Train Generator
        optimizer_g.zero_grad()
        output_fake = discriminator(fake_data).view(-1, 1)
        loss_g = criterion(output_fake, labels_real)
        loss_g.backward()
        optimizer_g.step()

        if i % 50 == 0:
            print(f"Epoch [{epoch}/{num_epochs}] Batch {i}/{len(dataloader)} \
                  Loss D: {loss_d.item()}, loss G: {loss_g.item()}")

    # Early Stopping Check
    current_loss = loss_d.item() + loss_g.item()
    if current_loss < best_loss:
        best_loss = current_loss
        trigger_times = 0
        # Save the models
        torch.save(generator.state_dict(), save_path_g)
        torch.save(discriminator.state_dict(), save_path_d)
        print(f"Models saved at epoch {epoch+1}")
    else:
        trigger_times += 1
        print(f"No improvement in loss for {trigger_times} epochs")
        if trigger_times >= patience:
            print(f"Early stopping at epoch {epoch+1}")
            break

# Save the final models in case the loop exits naturally
torch.save(generator.state_dict(), save_path_g)
torch.save(discriminator.state_dict(), save_path_d)
```
